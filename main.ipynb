Key points in the distribution graphs:
1. In `Temperature`, most of the values falls between 20°C - 30°C. Some of the readings rises upto 58°C also below 20°C. This histogram is has positive skewness(right-skewness).
2. In `Humidity`, most of the values fall between 50% - 80%. This histogram has a slight positive skewness. 
3. In `PM2.5`, most of the values falls between 0 µg/m³ - 15 µg/m³. There are also values greater than 100 µg/m³, but those are some rare cases. This is clearly a right-skewed(positive skewness) graph.
4. The `PM10` column is has more similar property with `PM2.5`. Here most of the values 10 µg/m³ - 25 µg/m³. Here too, there are rare occurance of values greater than 100 µg/m³. The Histogram shows a right-skewed distribution.
5. The column `NO2` has most of the values in the range of 20 ppb to 30 ppb. There are occasionally rare cases where the concentration has risen over 45 ppb. This Histogram shows a right-skewed distribution. 
6. The concentration of `SO2` lies mostly in the range of 5 ppb to 8 ppb. The rare events of concentration of SO2 occurs above 23 ppb. This Histogram shows right-skewed distribution.
7. In `CO` column, the data points are vastly distributed around 1.0 ppm. There are also some events where the concentration has reached over 3.0 ppm, but they are small in number. This histogram has also right-skewed distribution.
8. The column `Proximity_to_Industrial_Areas`, has the most values at 7 km and also between 10 to 12 km, meaning most of the data present in this dataset are recorded at 7km and 10-12 km from the industrial area.
9. The `Population_Density` is vastly distributed between 320 people/km² to 620 people/km². There are also datapoints where there are more than 900 people/km².

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
svm_f1 = f1_score(y_true=y_test, y_pred=svm_pred, average='micro')
svm_precision = precision_score(y_true=y_test, y_pred=svm_pred, average='micro')
svm_recall = recall_score(y_true=y_test, y_pred=svm_pred, average='micro')
svm_acc = accuracy_score(y_true=y_test, y_pred=svm_pred)
print(f"F1-score:", svm_f1)
print(f"Precision score:", svm_precision)
print(f"Recall Score:", svm_recall)
print(f"Accuracy:", svm_acc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true=y_test, y_pred=svm_pred))
```

Interpretation:
* Our model identified all the samples of class-0 correctly.
* It performs poor in predicting class-3

Metrics:
* Overall accuracy of 94% is really good.
* By analysing Macro average precision, our SVM model is precise by 0.92 in all of our classes, on average. Macro average recall of 0.91 gives that the model captures 91% of actual samples across all classes.
* Weighted average precision of 0.94 tells that 94% of the predicitions are correct. Weighted average Recall of 0.94 tells that 94% of the samples were correctly identified.

## Conclusion

```python
models = ['AdaBoostClassifier', 'DecisionTreeClassifier', 'ExtraTreeClassifier', 'GradientBoostingClassifier', 'NaiveBayes', 'RandomForestClassifier', 'SVC']
model_classes = [adaboost, tree, extra_tree, gb, nb, rf, svm]
result = {}

for model, name in zip(model_classes, models):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Get classification report as a dictionary
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # Initialize a dictionary for the current model
    result[name] = {"Accuracy": report["accuracy"]}
    
    # Add metrics for all classes dynamically
    for class_label in sorted(report.keys()):
        if class_label.isdigit():  # Only process numeric class labels
            result[name][f"Class {class_label} Recall"] = report[class_label]["recall"]
            result[name][f"Class {class_label} F1-Score"] = report[class_label]["f1-score"]
    
    # Add weighted metrics
    result[name]["Weighted Recall"] = report["weighted avg"]["recall"]
    result[name]["Weighted F1-Score"] = report["weighted avg"]["f1-score"]

# Convert the results dictionary to a pandas DataFrame
results_df = pd.DataFrame(result).T
results_df
```

* In terms of overall accuracy(overall performance of a model), Random Forest leads, but in learning the patterns for minority classes like class-2 and class-3, GradientBoosting is better.
* The precision and recall score for those classes are slightly higher compared to Random Forest. It is kind of a tie between Random Forest and Gradient Boosting. 
* Both models are well performing and maintains generalisation across all the classes.

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

models = [
    ("Support Vector Machine", svm),
    ("Naive Bayes", nb),
    ("Decision Tree", tree),
    ("Extra Tree Classifier", extra_tree),
    ("Random Forest", rf),
    ("AdaBoost Classifier", adaboost),
    ("Gradient Boosting", gb)
]

y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))

num_models = len(models)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  
axes = axes.flatten()  

for idx, (name, model) in enumerate(models):
    ax = axes[idx]
    
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)
    else:
        y_score = model.decision_function(X_test)
    
    for i in range(y_test_binarized.shape[1]):
        fpr, tpr, _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, label=f"Class {i} (AUC = {roc_auc:.2f})")
    
    ax.plot([0, 1], [0, 1], color="navy", linestyle="--", label="Random Chance")  
    ax.set_title(f"ROC Curve: {name}", fontsize=12)
    ax.set_xlabel("False Positive Rate", fontsize=10)
    ax.set_ylabel("True Positive Rate", fontsize=10)
    ax.legend(loc="lower right", fontsize=8)
    ax.grid()

# Hide unused subplots if the number of models is less than the grid size
for idx in range(len(models), len(axes)):
    fig.delaxes(axes[idx])
plt.tight_layout()
plt.show()
```
